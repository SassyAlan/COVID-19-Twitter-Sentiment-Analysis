# Data Cleaning

Data Collection Pipeline Explanation

The function get_tweet_ids takes in a file containing dates in the format of YYYY-MM-DD. The function then loops through all the dates and downloads the tweet IDs the Panacea Lab collected for that day. It then randomly selects 300 tweets without replacement and stores the IDs in a list. The file that is downloaded from the lab is then deleted. This process repeats for all 307 days we are interested in analyzing. At the end of the function, the tweet IDs are written into a txt file to easily pass it in to twarc.

The function rehydrate takes in a txt file containing the tweet IDs we wish to rehydrate. The txt file we saved from the previous function is passed in and the tweets are rehydrated and saved in a json file. This is done through twarc which calls on Twitter's API. The json file contains Tweet objects, which the documentation for can be found here, https://developer.twitter.com/en/docs/twitter-api/v1/data-dictionary/object-model/tweet.

```python
#Function to get tweet ids we wish to analyze and save it in a txt file
#Takes a while to run 10-15 mins
def get_tweet_ids(dates_file):
    
    #Opens the text file of dates that we wish to analyze
    #307 dates 
    text_file = open(dates_file, "r")
    dates = text_file.read().splitlines()
    text_file.close()
    
    #Should contain around 90,000 tweet IDs at the end
    ids_to_hydrate = []
    
    for i in range(len(dates)):
        
        current_date = dates[i]
        repo_link = "https://github.com/thepanacealab/covid19_twitter/raw/master/dailies/" + current_date + "/" + current_date + "_clean-dataset.tsv.gz"
        file_name = current_date + "_clean-dataset.tsv.gz"
        unzipped_file = current_date + "_clean-dataset.tsv"
        
        os.system("wget " + repo_link) #Get file from the Panacea Lab's repo
        os.system("gzip -d " + file_name) # Extract desired file and unzip it
        
        #Read in the tsv file containing the tweet IDs
        tweet_ids = pd.read_csv(unzipped_file, delimiter="\t")
        
        #Select the 300 tweets we wish to analyze for the current date
        #Stores it in a list that we append to our combined list of tweet ids
        one_day = list(tweet_ids['tweet_id'].sample(n=300, replace=False).values)
        ids_to_hydrate = ids_to_hydrate + one_day
        
        #Removes the tsv file after we're done sampling to reduce clutter
        os.remove(unzipped_file)
     
    #Writes the resulting list of 90,000 tweets to a txt file so we can use twarc to rehydrate these tweets
    with open("ids.txt", mode="w") as output_file:
        for ids in ids_to_hydrate:
            output_file.write("%s\n" % ids)
        
    return ids_to_hydrate
```

```python
#Rehydrates our tweet ids into json data, takes in the file that is saved from get_tweet_ids
#Takes about 20 minutes to rehydrate 90,000 tweets
def rehydrate(tweet_ids):
    #Calls on twarc
    os.system("twarc hydrate " + tweet_ids + " > tweets.json")
```

Tidy Data Format - json
- We receive our data in a .json format, which is a tidy data format.  Through the process of tidying it further we removed columns and rows which serve as no use to our analysis, this process and reasoning for doing so are below.

Why we reduced the amount of tweets (from 1000 to 300),
- Originally we wanted to analyze 1000 tweets per day from the date range of 2020-03-22 to 2021-01-22, however the size of the .json and dataframe that were generated from that loaded very slowly on our computers and was not practical.  In order to work more easily with the data, we reduced the number of tweets per day to 300 to speed up the process of loading/uploading the data.

Why we filtered out other languages besides English and Spanish,
- We are using a sentiment analysis API to analyze the sentiment of the tweets in our dataset, and the API only works on English and Spanish texts. On top of this, we are choosing to look mainly at political and current events in the United States, and English and Spanish are the two most common languages in the US by far.

Why we dropped certain columns,
- `__ID` and `_Str` attributes (e.g `quoted_status_id`, `quoted_status_id_str`, etc.) - unnecessary information
- `In_reply_to` attributes (e.g `status_id`, `status_id_str`, etc.) - outside the scope of our analysis
- Booleans - `truncated`, `possibly_sensitive` - all values False in these columns 
- `withheld_ attributes` → getting rid of the other two (`_scope`, and `_in_countries`) as it is unnecessary information, but planning to keep `withheld_copyright` since it contained a row with a “True” values so will filter out that row later on
- User attributes (`source`, `contributors`) - unnecessary information 

Why we kept certain columns
- `full_text` - gives us the actual tweet being made, which we are analyzing.
- `lang`, `entities`, `extended_entities`, `withheld_copyright` - Gives us additional information about the tweet that is not shown in the text.  This can be emojis, links, etc. which help us to better understand the sentiment. 
- `is_quote_status`, `retweet_count`, `favorite_count`, `favoirited`, `retweeted` - Also give us some insight into how popular the tweet may have been, giving us some idea of how other people may be feeling about the same idea / topic.
- `coordinates`, `geo`, `place`, `user` - contain user info (ex: location/coordinate info) that would be helpful to identify the region of the tweets.

```python
#Reads in the json file into a dataframe
df = pd.read_json("tweets.json", lines=True, orient="records")
df.head()
```

```python
def filterTweets(aDF):
    aDF = aDF.loc[(aDF['lang'] == 'en') | (aDF['lang'] == 'es') | (aDF['lang'] == 'und')]
    aDF = aDF[desiredColumns].reset_index(drop=True)
    return aDF

def dfToDictInChunks(df, chunkSize):
    dfList = [ df.loc[i : i + chunkSize - 1, :] for i in range(0, len(df), chunkSize)]
    dictList = []
    for slicedDF in dfList:  
        result = slicedDF.to_dict('records')
        dictList.append(result)
    return dictList

# Take an unformatted dictionary of tweets, and formats it to send to the api.
def formatDictList(unformattedDictList):
    formattedList = []
    for chunk in unformattedDictList:
        formattedDict = {}
        formattedDict['language'] = 'auto'
        formattedDict['data'] = chunk
        formattedList.append(formattedDict)
    
    return formattedList
```

```python
# Filter unwanted columns and rows from df.
df = filterTweets(df)
# Make subset of DF to send to sentiment API
tweetsDF = df[ ['id', 'full_text'] ].copy()
tweetsDF.columns = ['id', 'text']

# format tweets and chunk into 10,000 tweet sublits to send to api
unformattedTweetsList = dfToDictInChunks(tweetsDF, 10000)
formattedTweetsList = formatDictList(unformattedTweetsList)
```

```python
# Send tweets to sentiment api and add reponses to responseList
sentiment_url = 'http://www.sentiment140.com/api/bulkClassifyJson'

responseList = []
for tweetDict in formattedTweetsList:
    tempJSON = json.dumps(tweetDict)
    res = requests.post(sentiment_url, data=tempJSON)
    responseList.append(res)
```

Obtaining sentiment values for the text of each tweet in the dataset. Outline of Steps:

   1. First we filter out the unwannted columns from our dataframe, using the function filterTweets, which drops  and tweet that is not of a language english, spanish, undefined (which is typically emojis or links).
    
   2. Next we form a new DF "tweetsDF", which will be formatted specifically to send to the API Sentiment140.
    
   3. Sentiment 140 only needs the ID and full text of each row.  Also it can only take 10,000 tweets at a time, so it needs to be chunked into 10,000 tweet blocks.
    
   4. After sending the tweets to the API they needed to be grouped back into the main DF, which is done by merging the dfs on 'ID'.
    
   5. Now all of the tweets have an associated sentiment or polarity, and we write that df to the disk as a .json.

```python
# Clean up response into format that is easy to convert to JSON
text_polarity = {}
for response in responseList:
    for tweet in response.json()['data']:
        text_polarity[tweet['id']] = {
            'id' : tweet['id'],
            'text' : tweet['text'],
            'polarity' : tweet['polarity']
        }
```

```python
# Write new data to its own json file
outfile = open("jsonWithSentiment.json", "w")
outfile.write(json.dumps(text_polarity, indent=4))
outfile.close()
```

```python
# Convert json to DF
sentiment_df = pd.read_json(json.dumps(text_polarity), orient='index')
sentiment_df.head()
# Remove text row as it is redundent
reduced_df = sentiment_df[[ 'id', 'polarity' ]]
# Merge sentiment into main df, so that polarity is matched with its original tweet
df = pd.merge(df, reduced_df, on='id')
df.head()
```

```python
#standardize functions

def standardize_Covid (string):
    string = string.lower()
    string = string.strip()
    string =string.replace('covid_19', 'covid19')
    string =string.replace('covidー19', 'covid19')
    string =string.replace('covid__19', 'covid19')
    
    return string

def standardize_hashtags (string):
    string = string.lower()
    string = string.strip()
    string = string.replace('_',' ')
    string = string.replace('-',' ')
    string = string.replace('ー',' ')
    string = string.replace('vaccines','vaccine')
    
    if 'stayathome' in string:
        output = 'stayhome'
    elif 'quédateencasa' in string:
        output = 'quedateencasa'
    else:
        output = string
    
    output = string.strip()
    
    return output
```

```python
updated_text_hashtags = text_hashtags.apply(standardize_hashtags)
updated_text_hashtags = text_hashtags.apply(standardize_Covid)
df = df.assign(updated_hashtags = updated_text_hashtags)
print('number of hashtags used: ', text_hashtags.nunique())
# top 20 common hashtags after standardizing
print('top 20 common used hashtags after standardizing')
df.updated_hashtags.value_counts().head(20)
```

```python
nltk.download('stopwords')

def ngrams_cleaning(aDF):
    #Removes hashtags and emojis
    aDF['clean_text'] = aDF['full_text'].apply(lambda x: p.clean(x))

    #Removes punctuations and changes to lower case
    aDF['clean_text'] = aDF['clean_text'].apply(lambda x: re.sub(r'[^\w\s]', '', x))
    aDF['clean_text'] = aDF['clean_text'].apply(lambda x: x.lower())

    #Removes all the digits
    aDF['clean_text'] = aDF['clean_text'].apply(lambda x: re.sub('[0-9]', '', x))

    #Tokenizes the text to make it easier to analyze ngrams
    tokenizer = TweetTokenizer()
    aDF['clean_text'] = aDF['clean_text'].apply(lambda x: tokenizer.tokenize(x))

    #predefined stop words/common words
    stopwords_english = set(nltk.corpus.stopwords.words('english'))
    stopwords_spanish = set(nltk.corpus.stopwords.words('spanish'))
    #Removes all stopwords
    aDF['clean_text'] = aDF['clean_text'].apply(lambda x: [i for i in x if i not in stopwords_english])
    aDF['clean_text'] = aDF['clean_text'].apply(lambda x: [i for i in x if i not in stopwords_spanish])
    
    return aDF

def get_ngrams(aDF, n):
    return aDF['clean_text'].apply(lambda x: list(nltk.ngrams(x, n)))
```

##### Sentiment Across Time
To better see the sentiment trend over time, we ran a correlation test and a linear regression to see if sentiment is negatively correlated with time and that if time is a significant predictor for sentiment in tweets.

```python
#assign values to the date to better see the relationship trend between polarity and time 
df['date-int'] = pd.to_datetime(df['date']).dt.strftime('%Y%m%d').astype(int)
date_dict = df['date-int'].value_counts()
date_list = list(date_dict.keys())
date_list.sort()
date_int_map = {}

cur_int = 1
for date in date_list:
    if date not in date_int_map:
        date_int_map[date] = cur_int
        cur_int += 1

df['date_val'] = df['date-int'].apply(lambda x : date_int_map[x])
df['date_val']=df['date_val'].astype(np.int)
df.head()
```

```python
#Explore the relationship between time and sentiment in tweets through correlation coefficients
np.corrcoef(df['polarity'], df['date_val'])
```

```python
#Test whether time is a significant predictor for tweets sentiment
outcome, predictors = patsy.dmatrices('polarity ~ date_val ', df)
mod = sm.OLS(outcome, predictors)
res = mod.fit()
print(res.summary())
```

We found that sentiment is slightly negatively correlated with time, such that as time goes by, the sentiment in tweets became more negative, although not significant (r = -0.006). Moreover, in the linear regression model where time is the sole predictor of tweets sentiment, we found that time is not a significant predictor for sentiment in tweets (F(1,62389) = 2.21, p = 0.14).

##### Sentiment by Language

To better understand if language is a predictor for sentiment in the tweets, we ran an one-way ANOVA with language as the independent variable and sentiment polarity as the dependent variable; we also ran a post-hoc Tukey HSD test (to control for family wise error rate) to test pairwise comparisons.

Plotting the average sentiment of tweets separated between English and Spanish, we get a clear picture that Spanish tweet tended to be more positive than English. This gives us a good indication that language might be a significant predictor for the sentiment of tweets. Excluded 'und' (undefined) column because most of the text was emojis or links, and the polarity was neutral. 

```python
# Test whether language is a significant predictor for sentiment in a simple regression test
outcome, predictors = patsy.dmatrices('polarity ~ lang ', df)
mod = sm.OLS(outcome, predictors)
res = mod.fit()
print(res.summary())
```

```python
# get an ANOVA table and F statistics
import statsmodels.api as sm
from statsmodels.formula.api import ols
model = ols('polarity ~ lang', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

```python
# perform multiple pairwise comparison (Tukey HSD)
from statsmodels.stats.multicomp import pairwise_tukeyhsd
m_comp = pairwise_tukeyhsd(endog=df['polarity'], groups=df['lang'], alpha=0.05)
print(m_comp)
```

We can see that language is a significant predictor for tweet sentiments (F(2,62388) = 196.45, p<0.001). Specifically, Spanish tweets (M = 2.16) are significantly more postive than English tweets (M = 2.00; p = 0.001).

# Ethics & Privacy

The sentiment of a tweet is subjective; the algorithm which determines the sentiment may not always be accurate. The group of individuals who use twitter may fall into specific categories, i.e. age groups, political affiliations, etc., and this may bias the events which cause the sentiment of tweets to vary.  We will try to mitigate this bias by using random subsets of tweets, and therefore attempting to achieve the most diverse set of opinions as possible, however this is limited to the demographics of the twitter user base.

The data gathered for our project is within the public domain, and is available to anyone through twitter.  Our analysis will not include any information about users more than is available through their twitter profile.  We are obtaining the data through both the twitter api and apis managed by other groups.  Because of this we are not obtaining consent from any individuals to use their twitter information, however we will be obtaining consent from the owners of the apis we are using in order to gain access to their datasets. 

The data we are interested in, specifically the types of tweets that made it into our dataset, is outlined in the data cleaning section. We specifically altered the dataset into Tidy Data Format (json), reduced the number of tweets per day, filtered out language, dropped and kept certain columns from the twitter data directory. Ultimately, modifying our dataset to only include certain aspects from the initial directory will affect our analysis in that it narrows down what we are specifically looking to analyze, which is the sentiment of tweets relating to the COVID-19 pandemic. However, the process of filtering out the data to include certain types of tweets may overlook some basic information, leading to bias in our final analysis.  

# Conclusion & Discussion

We came into this analysis asking the following question: How has the sentiment of tweets been impacted and changed over the course of the COVID-19 pandemic when factoring in highly significant events that took place, specifically the emergence and approval of the COVID-19 vaccine, and language differences? 

Throughout our analysis, we observed that the results did not support our hypothesis which stated that the sentiment of tweets became more negative as a function of time. Through a bivariate correlation test, we found that sentiment of tweets is negatively correlated with time. However, in a subsequent linear regression test in which time was the independent variable and tweet sentiment was the dependent variable, we found that time was not a significant predictor for sentiment of tweets. 

Furthermore, the results did not support our hypothesis that on days where there were significant events, such as the vaccine announcements, the sentiment of tweets would show more significant positivity. On days with important vaccine announcements, the sentiment for these days were still similar to the overall sentiment of our dataset which is not what we expected. However, on these days there were clear increases in usage of the words “vaccines”, “first”, and “Dr.” in tweets, demonstrating that users were interested in these events. Interestingly, we did find a significant difference in the sentiment of English and Spanish tweets which was unexpected. Moreover, post-hoc pairwise comparisons tests suggest that Spanish tweets show significantly more positive sentiment than English tweets. Future research should look deeper into why this might have occurred.

Although the API we used to analyze the sentiment of our tweet dataset is typically considered one of the best APIs for analyzing tweets, it did seem to have some limitations in regards to our analysis.  First of all, because tweets are not a typical form of human language the API relies heavily on categorization of tweets to aid in analyzing their sentiment.  Additionally, our tweets were grouped together through the category of “Covid”, however this did not aid the API very much in terms of its analysis with categorized tweets and non-categorized tweets receiving the same evaluations.  We believed that because of this many tweets were labeled as neutral when they maybe should not have been, and many other tweets seems to have been miscategorized.
